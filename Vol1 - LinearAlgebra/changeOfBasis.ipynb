# Change Of Basis Exercises

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the movie vectors
movie_1 = np.array([2, 3])
movie_2 = np.array([1, 1])

# Define the transformation matrix
T = np.array([[1, -1],
              [2, 1]])
```

Use the provided transformation matrix to perform a basis change from the genre basis to the user preference basis.

```python
# Perform the basis change for each movie vector
transformed_movie_1 = np.dot(T, movie_1)
transformed_movie_2 = np.dot(T, movie_2)

# Output the transformed vectors
print("Transformed Movie_1 Vector:", transformed_movie_1)
print("Transformed Movie_2 Vector:", transformed_movie_2)
```

Plot the original movie vectors and the transformed movie vectors on the same graph and use different colors or
markers to distinguish between the original and transformed vectors.

```python
# Plotting the original movie vectors and transformed movie vectors on the same graph
plt.quiver(0, 0, *movie_1, scale=1, scale_units='xy', angles='xy', color='r', label="Movie 1 - Original")
plt.quiver(0, 0, *movie_2, scale=1, scale_units='xy', angles='xy', color='b', label="Movie 2 - Original")
plt.quiver(0, 0, *transformed_movie_1, scale=1, scale_units='xy', angles='xy', color='g', linestyle='dashed', label="Movie 1 - Transformed")
plt.quiver(0, 0, *transformed_movie_2, scale=1, scale_units='xy', angles='xy', color='y', linestyle='dashed', label="Movie 2 - Transformed")

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend(loc='upper left', bbox_to_anchor=(0, 1))
plt.show()
```

Check the linear independence of the transformed movie vectors in the new basis and interpret the results
in the context of user preferences.

```python
# Check linear independence of the transformed vectors by examining the rank of the matrix formed by the vectors
transformed_matrix = np.vstack((transformed_movie_1, transformed_movie_2))
rank = np.linalg.matrix_rank(transformed_matrix)

# If rank is equal to the number of vectors, then the vectors are linearly independent.
if rank == len(transformed_matrix):
    print("The transformed movie vectors are linearly independent.")
else:
    print("The transformed movie vectors are linearly dependent.")
```
